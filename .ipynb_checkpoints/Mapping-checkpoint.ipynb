```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from PIL import Image
```

```python
image = Image.open("/Users/millie/downloads/20181129_132810.jpg")
```

```python
plt.imshow(image)
```

```python
from PIL.ExifTags import TAGS, GPSTAGS
```

```python
info = image._getexif()
```

```python
def gps_find(info):
    positions = [info[34853][2], info[34853][4]]
    coordinates = []
    for pos in positions:
        gps_coord = pos[0][0]/pos[0][1] + pos[1][0]/(pos[1][1]*60) + pos[2][0]/(pos[2][1]*3600)
        coordinates.append(gps_coord)
    return pd.DataFrame([{'latitude':coordinates[0], 'longitude':coordinates[1]}])
```

```python
coordinates = gps_find(info)
```

```python
coordinates
```

```python
fig = plt.gcf()
fig.set_size_inches(8, 6.5)

map = Basemap(llcrnrlon=-110,llcrnrlat=36.5,urcrnrlon=-101.,urcrnrlat= 41.5,
             resolution='l', projection='cass', lat_0=39,lon_0=-107)


lons = coordinates.latitude.values
lats = coordinates.longitude.values

plt.figure(figsize=(10,10))
map.shadedrelief()
map.drawstates(color='0.2')

x, y = map(lons, lats)
x, y = map(-107, 39)
map.scatter(x, y, marker='D',color='m')

x, y = map(lons, lats)

map.scatter(x, y, 10, marker='o', color='Red') 


plt.show()
```
